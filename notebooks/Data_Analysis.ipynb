```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('D:\GooglePlaystore\data\googleplaystore_cleaned.csv')
```

```python
df.head(5)
```

Average rating of apps on Playstore?

```python
avg_rating= df['Rating'].mean()
print(f'Average Rating of apps on Playstore is : {avg_rating:.2f}')
```

Percentage of apps are free vs paid?

```python
total= df['Type'].shape[0]
free= df[df['Type']=='Free'].shape[0]
paid= df[df['Type']=='Paid'].shape[0]

percentfree= (free/total) * 100
percentpaid= (paid/total) * 100

print(f'Free apps are {percentfree:.2f}%')
print(f'Paid apps are {percentpaid:.2f}%')
```

Most common app category?

```python
app_category= df['Category'].value_counts().head(10)
common_app= app_category.index[0]
high_number= app_category.values[0]
print(f'The most common app category is {common_app}')
print(f'The {common_app} app category has highest number of app which is {high_number}')
```

Highest number of apps category wise.

```python
plt.figure(figsize=(10, 5))
sns.barplot(x= app_category.index, y= app_category.values, palette= 'dark')
plt.xticks(rotation= 45)
plt.xlabel('Category')
plt.ylabel('Number of apps')
plt.title('Highest number of apps in top 10 category')
plt.show()
```

Distribution of content rating?

```python
content_rating= df['Content Rating'].value_counts()
plt.figure(figsize=(10, 5))
sns.barplot(x= content_rating.index, y= content_rating.values, palette= 'dark')
plt.xlabel('Content Rating')
plt.ylabel('Number of apps')
plt.show()
```

Total apps that belong to the top 5 most popular category?

```python
category= df['Category'].value_counts().head(5)

sns.barplot(x= category.index, y= category.values, palette= 'dark')
plt.show()
print(f'Total number of apps in top 5 category is {category.values.sum()}')
```

Most common rating given to apps?

```python
sns.histplot(df['Rating'], bins= 20)
plt.title('Most common rating given to app')
plt.xlabel('Ratings')
plt.ylabel('Numbe of apps')
plt.show()
```

App prices distributed?

```python
plt.subplot(1,2,1)
sns.distplot(df['Price'])
plt.subplot(1,2,2)
sns.boxplot(df['Price'])
plt.ylabel('Prices')
plt.title('App prices distributed')
```

Free apps ratings vs paid apps ratings?

```python
free_paid_rating= df[['Type', 'Rating']].groupby('Type').mean().reset_index()

sns.barplot(x= free_paid_rating['Type'], y= free_paid_rating['Rating'], palette= 'dark')
```

Highest ratings of app categories?

```python
avg_rating= df[['Category', 'Rating']].groupby('Category').mean().sort_values(by= 'Rating', ascending= False).reset_index().head(10)
avg_rating
```

Corelation between installs and ratings of apps?

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x= df['Installs'], y= df['Rating'], alpha= 0.5)
plt.title('Installs Vs Ratings')
plt.xlabel('Number of installs')
plt.ylabel('App ratings')
plt.xscale('log')
plt.show()
```

Paid vs free apps intalls?

```python
paid_vs_free_installs= df[['Type', 'Installs']].groupby('Type')[['Installs']].mean().reset_index()
print(f'Average installs of free app : {paid_vs_free_installs['Installs'][0]:.2f}')
print(f'Average installs of paid app : {paid_vs_free_installs['Installs'][1]:.2f}')
```

Expensive vs free apps rating?

```python
expen_vs_free= df[['Price', 'Rating']].groupby('Price')[['Rating']].mean().reset_index().sort_values(by='Rating', ascending= False)
plt.figure(figsize= (10,5))
sns.barplot(x= expen_vs_free['Price'], y= expen_vs_free['Rating'], palette= 'dark')
plt.title('Average rating of prices')
```

Most paid app category?

```python
most_paid= df[df['Type']=='Paid'].groupby('Category')[['Price']].count().sort_values(by= 'Price', ascending= False).reset_index().head(10)
plt.figure(figsize= (10, 5))
sns.barplot(x= most_paid['Category'], y= most_paid['Price'])
plt.xticks(rotation= 70)
```
